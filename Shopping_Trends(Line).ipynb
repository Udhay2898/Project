```python
for column in df.columns:
    unique_values = df[column].unique()
    print(f"{column}: {unique_values}\n")
```

```python
agg_data = df.groupby('Season').agg({
    'Promo Code Used': 'count',
    'Subscription Status' : 'count'}).reset_index()
agg_data
```

```python
agg_data.plot(figsize = (20,6),marker = 'o', grid = True)
plt.show()
```

```python
Season_data = df.groupby(['Season', 'Gender']).agg({'Purchase Amount (USD)': 'sum'}).reset_index()
Season_data
```

```python
Season_pivot = Season_data.pivot(index='Season', columns='Gender', values='Purchase Amount (USD)')
Season_pivot
```

```python
Season_pivot.plot(figsize = (20,6),marker = 'o',grid = True)
plt.show()
```

```python
grouped_data = df.groupby(['Payment Method', 'Season']).agg({'Purchase Amount (USD)': 'sum'}).reset_index()
grouped_data
```

```python
Payment_pivot_data = grouped_data.pivot(index='Payment Method', columns='Season', values='Purchase Amount (USD)')
Payment_pivot_data
```

```python
Payment_pivot_data.plot(figsize = (16,6),marker = 'o',grid = True,title = 'Payment Method by Season')
plt.show()
```

```python
agg_data = df.groupby(['Season', 'Gender', 'Size'])['Purchase Amount (USD)'].sum().unstack(['Gender', 'Size'], fill_value=0)
agg_data
```

```python
agg_data.plot(figsize = (16,8), marker ='o',grid = True,title = 'Size and Gender wise Purchase Amount')
plt.show()
```
