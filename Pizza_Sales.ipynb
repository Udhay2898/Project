```python
agg_data = df.groupby('pizza_name').agg({
    'quantity': 'sum',
    'total_price': 'sum'}).reset_index()

agg_data.head(5)
```

```python
pizza_name = agg_data['pizza_name']
values = {
    'quantity' : agg_data['quantity'],
    'total_price' : agg_data['total_price']
}

bar_width = 0.2
index = range(len(pizza_name))

fig, ax = plt.subplots()

for i, (label, data) in enumerate(values.items()):
    plt.bar([x + i * bar_width for x in index], data, bar_width, label=label)

plt.xlabel('Pizza Name')
plt.ylabel('Count')
plt.title('Sum of Quntity and Total Price')
plt.xticks([x + bar_width for x in index],pizza_name,rotation = 90)
plt.legend()

plt.show()
```
