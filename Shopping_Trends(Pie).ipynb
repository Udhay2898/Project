```python
# Aggregate item counts
item_counts = df['Item Purchased'].value_counts()

# Plot pie chart
plt.figure(figsize=(10, 7))
plt.pie(item_counts, labels=item_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Items Purchased Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
# Create pie charts for each categorical variable

# Gender distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()

# Category distribution
category_counts = df['Category'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Category Distribution')
plt.axis('equal')
plt.show()

# Subscription Status distribution
subscription_counts = df['Subscription Status'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(subscription_counts, labels=subscription_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Subscription Status Distribution')
plt.axis('equal')
plt.show()

# Shipping Type distribution
shipping_counts = df['Shipping Type'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(shipping_counts, labels=shipping_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Shipping Type Distribution')
plt.axis('equal')
plt.show()

# Season distribution
season_counts = df['Season'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(season_counts, labels=season_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Season Distribution')
plt.axis('equal')
plt.show()

# Size distribution
size_counts = df['Size'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(size_counts, labels=size_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Size Distribution')
plt.axis('equal')
plt.show()

# Payment Method distribution
payment_counts = df['Payment Method'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Payment Method Distribution')
plt.axis('equal')
plt.show()


# Color distribution
color_counts = df['Color'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(color_counts, labels=color_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Color Distribution')
plt.axis('equal')
plt.show()

# Frequency of Purchases distribution
frequency_counts = df['Frequency of Purchases'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(frequency_counts, labels=frequency_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Frequency of Purchases Distribution')
plt.axis('equal')
plt.show()

# Location distribution
location_counts = df['Location'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(location_counts, labels=location_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Location Distribution')
plt.axis('equal')
plt.show()
```

```python
# Separate data by gender
male_data = df[df['Gender'] == 'Male']
female_data = df[df['Gender'] == 'Female']

# Aggregate item counts for each gender
male_item_counts = male_data['Item Purchased'].value_counts()
female_item_counts = female_data['Item Purchased'].value_counts()

# Plotting pie charts for each gender
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart for male
ax1.pie(male_item_counts, labels=male_item_counts.index, autopct='%1.1f%%', startangle=140)
ax1.set_title('Items Purchased by Males')
ax1.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.

# Pie chart for female
ax2.pie(female_item_counts, labels=female_item_counts.index, autopct='%1.1f%%', startangle=140)
ax2.set_title('Items Purchased by Females')
ax2.axis('equal')

plt.tight_layout()
plt.show()
```

```python
# Separate data by gender and category
male_clothing = df[(df['Gender'] == 'Male') & (df['Category'] == 'Clothing')]
male_footwear = df[(df['Gender'] == 'Male') & (df['Category'] == 'Footwear')]
female_clothing = df[(df['Gender'] == 'Female') & (df['Category'] == 'Clothing')]
female_footwear = df[(df['Gender'] == 'Female') & (df['Category'] == 'Footwear')]

# Aggregate item counts for each gender and category
male_clothing_counts = male_clothing['Item Purchased'].value_counts()
male_footwear_counts = male_footwear['Item Purchased'].value_counts()
female_clothing_counts = female_clothing['Item Purchased'].value_counts()
female_footwear_counts = female_footwear['Item Purchased'].value_counts()

# Plotting pie charts for each gender and category
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Pie chart for Male Clothing
axes[0, 0].pie(male_clothing_counts, labels=male_clothing_counts.index, autopct='%1.1f%%', startangle=140)
axes[0, 0].set_title('Male Clothing')

# Pie chart for Male Footwear
axes[0, 1].pie(male_footwear_counts, labels=male_footwear_counts.index, autopct='%1.1f%%', startangle=140)
axes[0, 1].set_title('Male Footwear')

# Pie chart for Female Clothing
axes[1, 0].pie(female_clothing_counts, labels=female_clothing_counts.index, autopct='%1.1f%%', startangle=140)
axes[1, 0].set_title('Female Clothing')

# Pie chart for Female Footwear
axes[1, 1].pie(female_footwear_counts, labels=female_footwear_counts.index, autopct='%1.1f%%', startangle=140)
axes[1, 1].set_title('Female Footwear')

plt.tight_layout()
plt.show()
```

```python
# Define a function to categorize 'Review Rating'
def categorize_rating(rating):
    if rating == 5:
        return 'Excellent'
    elif rating >= 4.0:
        return 'Good'
    elif rating >= 3.0:
        return 'Above Average'
    elif rating >= 2.0:
        return 'Average'
    elif rating >= 1.0:
        return 'Below Average'
    else:
        return 'Poor'

# Apply the function to create a new column 'Rating Category'
df['Rating Category'] = df['Review Rating'].apply(categorize_rating)
df.head()
```
