```python
agg_data = df.groupby('Season').agg({
    'Promo Code Used': 'count',
    'Subscription Status' : 'count'}).reset_index()

agg_data.head(5)
```

```python
Season = agg_data['Season']
values = {
    'Promo Code Used' : agg_data['Promo Code Used'],
    'Subscription Status' : agg_data['Subscription Status'],
}

bar_width = 0.4
gab = 2
index = range(len(Season))

fig, ax = plt.subplots(figsize=(12, 8))

for i, (label, data) in enumerate(values.items()):
    plt.bar([x * gab + i * bar_width for x in index], data, bar_width, label=label)

plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Count')
plt.xticks([x * gab + (len(values) - 1) * bar_width / 2 for x in index], Season, rotation=0)

plt.grid()
plt.legend()
plt.tight_layout()

plt.show()
```

```python
g = df.groupby('Location')
Location_df = {}
```

```python
for location, group in g:
    agg_df = group.groupby('Season').agg({
        'Purchase Amount (USD)': 'sum',
        'Previous Purchases': 'sum'
    }).reset_index()
    Location_df[location] = agg_df
```

```python
bar_width = 0.4
fig, axes = plt.subplots(nrows=len(Location_df), ncols=1, figsize=(10, 7 * len(Location_df)))


for i, (location, location_df) in enumerate(Location_df.items()):
    ax = axes[i]
    location_df.plot(kind='bar', x='Season', y=['Purchase Amount (USD)', 'Previous Purchases'], ax=ax, width=bar_width)
    ax.set_title(location)
    ax.set_xlabel('Season')
    ax.set_ylabel('Count')
    ax.legend(['Purchase Amount (USD)', 'Previous Purchases'])
    ax.grid()
    ax.set_xticklabels(location_df['Season'], rotation=0)

plt.tight_layout()
plt.show()
```

```python
agg_data = df.groupby(['Season', 'Gender', 'Size'])['Purchase Amount (USD)'].sum().unstack(['Gender', 'Size'], fill_value=0)

# Plotting the bar chart
bar_width = 0.15
gap = 0.1  # Gap between the sets
seasons = agg_data.index
index = np.arange(len(seasons))

fig, ax = plt.subplots(figsize=(20, 10))

# Plotting bars for each combination of Gender and Size
num_bars = len(agg_data.columns)
total_width = num_bars * bar_width + (num_bars - 1) * gap
positions = [index + i * (bar_width + gap) for i in range(num_bars)]

for i, (gender, size) in enumerate(agg_data.columns):
    plt.bar(positions[i], agg_data[(gender, size)], bar_width, label=f'{gender} - {size}')

# Setting labels and titles
plt.xlabel('Season')
plt.ylabel('Purchase Amount (USD)')
plt.title('Purchase Amount by Season, Gender, and Size')

# Setting x-ticks and labels with gaps between sets
plt.xticks(index + (total_width - bar_width) / 2, seasons, rotation=0)

# Adding grid
plt.grid(True)

# Adding legend
plt.legend()

# Showing the plot
plt.show()
```
